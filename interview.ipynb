```python
#create client server communication
#server file
#below steps to establish TCP connection
import socket
#needs to establish connection
#in which ip address you will create server
#step 1:
s= socket.socket() #create socket
host = socket.gethostname() #to take my ip address use this funcion(gethostname)
port = 1234
#to create a communication we need host and port
#step 2
#bind host and port
s.bind((host, port)) #binding a host , port using a bind keyword in a socket 's' inside the tuple (bcz it is immutable and fast) 
#step 3
s.listen(5) #create time to listen for any connection from client using listen keyword in binded socket
#step 4
#accept the connection
while True: #use while for continous running
  conn, addr = s.accept() #It collects the address and connection of client in the  variable of conn, addr and it stores the connection address in socket 's', by accepting 
  print("Got Connection from : ", addr) 
  conn.send(b'Thank you connecting') #telling info to the client, 'b' is byte,bcz message will send in bytes
  conn.close()
________________________________________________________________________________________________________

#create client file
import socket
#1 Create port and socket, host
s= socket.socket()
host=socket.gethostname()
port = 1234
#2 Establish connection
s.connect((host, port)) #use connect keyword to connect to client
print(s.recv(1024))  #use recv keyword to receive the message and allocate the size of input
s.close()

 #how to use:
 #save files and open two cmd. types cd socket  in both cmd. first run server file. and run client program. 
```
